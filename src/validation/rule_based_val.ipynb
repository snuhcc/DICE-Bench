```python
import json
from pathlib import Path
```

# AI Call

```python
data_path = Path("../../outputs")
for data in data_path.glob("round_*.json"):
    if "scores" not in data.stem:
        if data.stem == "round_4":
            with open(data, "r", encoding="utf-8") as f:
                instance = json.load(f)
            for diag in instance:
                messages = diag["messages"]

                for round_dict in messages:
                    whole_messages = ""
                    for round_name, round_messages in round_dict.items():
                        for message in round_messages:
                            whole_messages += message["message"]
                    if "AI" not in whole_messages:
                        print(diag["metadata"]["diag_id"], round_name)
                        break
```

# Score-Based

```python
with open("../../dataset/archived/round_1_scores.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# 체크해야 할 가능한 키들
possible_keys = [
    "Persuassion_Deliberation_and_Negotiation",
    "Inquiry_and_Information_Seeking",
    "Eristic",
]

type_mismatch = [
    item for item in data if any(item["scores"].get(k) == 0 for k in possible_keys)
]

score_below = [item for item in data if item["scores"]["avg_score"] < 4]

print(f"type_mismatch: {len(type_mismatch)}")
print(f"score_below: {len(score_below)}")

common_items = [item for item in type_mismatch if item in score_below]
print(len(common_items))
```

# FilerData

```python
import json

# 1) JSON 파일 불러오기
with open("../../dataset/round_1.json", "r", encoding="utf-8") as f:
    data1 = json.load(f)

with open("../../dataset/round_2.json", "r", encoding="utf-8") as f:
    data2 = json.load(f)

with open("../../dataset/round_3.json", "r", encoding="utf-8") as f:
    data3 = json.load(f)

with open("../../dataset/round_4.json", "r", encoding="utf-8") as f:
    data4 = json.load(f)


# 2) 제거할 diag_id 목록 정의
rnd1_to_remove = [53, 410, 430]
rnd2_to_remove = [149, 199, 242, 250, 363, 415]
rnd3_to_remove = [
    13,
    29,
    67,
    111,
    143,
    167,
    173,
    189,
    216,
    238,
    248,
    255,
    320,
    350,
    404,
    433,
    446,
]
rnd4_to_remove = [37, 104, 115, 144, 214, 410, 415, 420]


# 3) 필터링(제거 대상이 아닌 데이터만 남김)
filtered_rnd1 = [
    item for item in data1 if item["metadata"]["diag_id"] not in rnd1_to_remove
]
filtered_rnd2 = [
    item for item in data2 if item["metadata"]["diag_id"] not in rnd2_to_remove
]
filtered_rnd3 = [
    item for item in data3 if item["metadata"]["diag_id"] not in rnd3_to_remove
]
filtered_rnd4 = [
    item for item in data4 if item["metadata"]["diag_id"] not in rnd4_to_remove
]


# 4) 필터링된 결과를 각각 새로운 JSON 파일로 저장
with open("../../dataset/round_1_filtered.json", "w", encoding="utf-8") as f:
    json.dump(filtered_rnd1, f, ensure_ascii=False, indent=4)

with open("../../dataset/round_2_filtered.json", "w", encoding="utf-8") as f:
    json.dump(filtered_rnd2, f, ensure_ascii=False, indent=4)

with open("../../dataset/round_3_filtered.json", "w", encoding="utf-8") as f:
    json.dump(filtered_rnd3, f, ensure_ascii=False, indent=4)

with open("../../dataset/round_4_filtered.json", "w", encoding="utf-8") as f:
    json.dump(filtered_rnd4, f, ensure_ascii=False, indent=4)
```
